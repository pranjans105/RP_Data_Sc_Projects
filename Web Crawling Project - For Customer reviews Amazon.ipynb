```python
# to get the top 5 customer reviews:

import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# website will be saved in this variable :
link = 'https://www.amazon.in/product-reviews/B089MV3MV7/ref=cm_cr_arp_d_viewpnt_lft?ie=UTF8&filterByStar=positive&reviewerType=all_reviews&pageNumber=1#reviews-filter-bar'
```

```python
# request the required details on website :
page = requests.get(link)
```

```python
# to get the content :
page
```

```python
page.content
```

```python
# Soup Parsing the content from website :
soup = bs(page.content, 'html.parser')
soup
```

```python
Names = soup.find_all('span',{'class': 'a-profile-name'})
```

```python
Names
```

```python
# Creating container for Names
cust_name = []
for i in range(0,len(Names)):
    cust_name.append(Names[i].get_text())
    
```

```python
cust_name
```

```python
# to see the stars of rating :
ratings = soup.find_all('span',{'class': 'a-icon-alt'})
```

```python
ratings.pop(12)
```

```python
ratings
```

```python
rate = []
for i in range(0,len(ratings)):
    rate.append(ratings[i].get_text())
```

```python
rate.pop(15)
```

```python
title = soup.find_all('a',{'data-hook':'review-title'})
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
```

```python
review_title
```

```python
review_title[:] = [title.rstrip('\n')for title in review_title]
```

```python
review_title
```

```python
# to remove '\n' from left side using lstrip
review_title[:] = [title.lstrip('\n')for title in review_title]
```

```python
review_title
```

```python
content = soup.find_all('span',{'class':'a-size-base review-text review-text-content'})
```

```python
content
```

```python
review_content = []
for i in range(0,len(content)):
    review_content.append(content[i].get_text)
```
