```python
# import required packages:
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
# Declaring the list of emty variables so that we can append the data overall 

na = []
ra = []
pr = []
```

```python
# Creating an array values and passing it into variables :
pages = np.arange(1,10,1)
```

```python
for page in pages:
    page = requests.get(url = "https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Mi&page="+str(page)+"&")
    soup = bs(page.text, 'html.parser')
    Phone_data = soup.findAll('div', {'class' : '_3pLy-c row'})
    sleep(randint(2,8))
```

```python
for store in Phone_data:
    name = store.find('div', {'class': '_4rR01T'})
    name.text.replace('\n', '')
    na.append(name)
    
    rate = store.find('span', {'class': '_2_R_DZ'})
    rate.text.replace('\n','')
    ra.append(rate)
    
    price = store.find('div', {'class': '_30jeq3 _1_WHN1'})
    price.text
    pr.append(price)
```

```python
MI_phone = pd.DataFrame({'Name': na, 'Rating': ra, 'Price': pr })
```
