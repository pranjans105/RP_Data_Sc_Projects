## Project 1: Product review of MI Phones for 6GB and above

```python
# import required packages:
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

### Project 1: To pull out the data of MI phones prices, review & ratings for analysis

```python
# Create variable for storing the information from link(website):
url = 'https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy%2C4io&otracker=nmenu_sub_Electronics_0_Mi'
```

```python
# Request from server and parse it into html
req = requests.get(url)
```

```python
# to parse the content from source website :
content = bs(req.content,'html.parser')
```

```python
content
```

```python
# Parse the source page/location which is required:
product_name = content.find_all('div',{'class': '_4rR01T'})
product_name
```

```python
print(product_name[0].text)
```

```python
len(product_name)
```

```python
# to print the ratings and reviews to print the data 
product_rating = content.find_all('span',{'class': '_2_R_DZ'})
```

```python
print(product_rating[0].text)
```

```python
len(product_rating)
```

```python
# to pull the prices of the products from source website :
product_price = content.find_all('div', {'class': '_25b18c'})
```

```python
print(product_price[0].text)
```

```python
len(product_price)
```

```python
# Now prepare 
print(product_name[0].text)
print(product_rating[0].text)
print(product_price[0].text)
```

```python
# Prepare 3 empty list and store the data in this list :

na = []
ra = []
pr = []

for i in product_name:
    na.append(i.text)

for i in product_rating:
    ra.append(i.text)
    
for i in product_price:
    pr.append(i.text)
```

```python
# Store data Name list, Price list, Rating list in DataFrame:
data = {'Name': na, 'Rating': ra, 'Price': pr}
```

```python
data = list(zip(na,ra,pr))
data
```

```python
dataf = pd.DataFrame(data, columns=['Product Name', 'Review & Ratings', 'Pricing'])
```

```python
dataf
```

```python
dataf.to_csv(r'MIphones.csv')
```
